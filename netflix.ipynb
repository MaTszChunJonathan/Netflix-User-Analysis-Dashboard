```python
import pandas as pd
netflix = pd.read_csv("Netflix Userbase.csv")
netflix
```

```python
netflix1 = netflix.copy()
netflix1 = netflix1[["User ID","Subscription Type","Monthly Revenue","Join Date","Last Payment Date","Country","Age","Gender"]]
netflix1
```

```python
netflix1["Join Date"] = pd.to_datetime(netflix1["Join Date"],dayfirst=True)
netflix1["Last Payment Date"] = pd.to_datetime(netflix1["Last Payment Date"],dayfirst=True)
netflix1["Months Joined"] = netflix1["Last Payment Date"].dt.to_period('M').astype(int) - netflix1["Join Date"].dt.to_period('M').astype(int)
netflix1["Total Revenue"] = netflix1["Monthly Revenue"] * netflix1["Months Joined"]
netflix1
```

```python
netflix1.to_excel('netflix_final.xlsx', sheet_name='Data')
```
